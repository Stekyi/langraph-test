```python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END

```

```python
class LifeTree(TypedDict):
    baby : str
    child: str
    teenage: str
    adult: str
    oldPerson: str

def babyNode(state: LifeTree) -> LifeTree:
    state['baby'] = 'you are born as a baby'
    return state

def childNode(state: LifeTree) -> LifeTree:
    state['child'] = 'you stop wearing dipers'
    return state

def teenageNode(state: LifeTree) -> LifeTree:
    state['teenage'] = 'you start developing your organs to aid u replace yourself'
    return state

def adultNode(state: LifeTree) -> LifeTree:
    state['adult'] = 'you move away from you parents and start life'
    return state

def oldPersonNode(state: LifeTree)-> LifeTree:
    state['oldPerson'] = 'You prepare to bid farewall to the world'
    return state


builder = StateGraph(LifeTree)

builder.add_node('baby_node', babyNode)
builder.add_node('child_node', childNode)
builder.add_node('teenage_node', teenageNode)
builder.add_node('adult_node', adultNode)
builder.add_node('oldPerson_node', oldPersonNode)

builder.add_edge(START, 'baby_node')
builder.add_edge('baby_node', 'child_node')
builder.add_edge('child_node', 'teenage_node')
builder.add_edge('teenage_node', 'adult_node')
builder.add_edge('adult_node', 'oldPerson_node')
builder.add_edge('oldPerson_node', END)

lifegraph = builder.compile()
```

```python
from IPython.display import Image, display
display(Image(lifegraph.get_graph().draw_mermaid_png()))
```
